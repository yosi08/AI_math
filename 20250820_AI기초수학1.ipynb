#2. 코랩 단축키


| 기능              | 단축키 (Windows/Linux)      | 단축키 (Mac)               |
| --------------- | ------------------------ | ----------------------- |
| **셀 실행**        | `Shift + Enter`          | `Shift + Enter`         |
| 셀 실행 + 아래로 이동   | `Alt + Enter`            | `Option + Enter`        |
| 셀 실행 + 위에 셀 추가  | `Ctrl + M` → `A`         | `Cmd + M` → `A`         |
| 셀 실행 + 아래에 셀 추가 | `Ctrl + M` → `B`         | `Cmd + M` → `B`         |
| 코드 셀 → 텍스트 셀 변경 | `Ctrl + M` → `M`         | `Cmd + M` → `M`         |
| 텍스트 셀 → 코드 셀 변경 | `Ctrl + M` → `Y`         | `Cmd + M` → `Y`         |
| 셀 삭제            | `Ctrl + M` → `D` (2번)    | `Cmd + M` → `D` (2번)    |
| 셀 복사            | `Ctrl + M` → `C`         | `Cmd + M` → `C`         |
| 셀 붙여넣기          | `Ctrl + M` → `V`         | `Cmd + M` → `V`         |
| 셀 위로 이동         | `Ctrl + M` → `K`         | `Cmd + M` → `K`         |
| 셀 아래로 이동        | `Ctrl + M` → `J`         | `Cmd + M` → `J`         |
| 셀 접기/펼치기        | `Ctrl + M` → `O`         | `Cmd + M` → `O`         |
| 전체 셀 접기         | `Ctrl + M` → `Z`         | `Cmd + M` → `Z`         |
| 전체 셀 펼치기        | `Ctrl + M` → `Shift + Z` | `Cmd + M` → `Shift + Z` |
| 모든 셀 다시 실행      | `Ctrl + F9`              | `Cmd + F9`              |
| 파일 저장           | `Ctrl + S`               | `Cmd + S`               |

#3. 마크다운 문법 정리

| 기능          | 문법 예시                                     | 결과 설명             |    |        |     |     |        |   |   |    |        |
| ----------- | ----------------------------------------- | ----------------- | -- | ------ | --- | --- | ------ | - | - | -- | ------ |
| 제목 (h1\~h6) | `# 제목1` \~ `###### 제목6`                   | 제목 크기를 6단계로 표현    |    |        |     |     |        |   |   |    |        |
| 굵게          | `**텍스트**`                                 | **굵은 글씨**         |    |        |     |     |        |   |   |    |        |
| 기울임         | `*텍스트*`                                   | *기울여진 글씨*         |    |        |     |     |        |   |   |    |        |
| 굵고 기울임      | `***텍스트***`                               | ***굵고 기울여진 글씨***  |    |        |     |     |        |   |   |    |        |
| 취소선         | `~~텍스트~~`                                 | ~~중간에 선이 그어진 글씨~~ |    |        |     |     |        |   |   |    |        |
| 인용문         | `> 인용할 문장`                                | 들여쓰기된 인용문         |    |        |     |     |        |   |   |    |        |
| 순서 있는 목록    | `1. 항목`                                   | 번호 있는 리스트         |    |        |     |     |        |   |   |    |        |
| 순서 없는 목록    | `- 항목` 또는 `* 항목`                          | 점 또는 별 리스트        |    |        |     |     |        |   |   |    |        |
| 링크          | `[텍스트](https://example.com)`              | 하이퍼링크 생성          |    |        |     |     |        |   |   |    |        |
| 이미지         | `![설명](이미지링크)`                            | 이미지 삽입            |    |        |     |     |        |   |   |    |        |
| 인라인 코드      | `` `코드` ``                                | `코드`              |    |        |     |     |        |   |   |    |        |
| 코드 블럭       | <pre>`python<br>print("Hello")<br>`</pre> | 여러 줄 코드 강조        |    |        |     |     |        |   |   |    |        |
| 수평선         | `---` 또는 `***`                            | 구분선 생성            |    |        |     |     |        |   |   |    |        |
| 표           | \`                                        | 헤더                | 헤더 | `<br>` | --- | --- | `<br>` | 값 | 값 | \` | 테이블 구조 |
| 수식 (LaTeX)  | `$x^2 + y^2$`<br>`$$x^2 + y^2 = z^2$$`    | 수학 수식 표현          |    |        |     |     |        |   |   |    |        |
| 체크박스 목록     | `- [ ] 할 일`<br>`- [x] 완료`                 | ✅ 체크리스트           |    |        |     |     |        |   |   |    |        |
